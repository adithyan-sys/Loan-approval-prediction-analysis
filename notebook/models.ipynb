```python
main_df['LoanAmount'] = main_df['LoanAmount'].fillna(main_df['LoanAmount'].median())
main_df[ 'Loan_Amount_Term'] = main_df['Loan_Amount_Term'].fillna(main_df['Loan_Amount_Term'].mean())
main_df['Credit_History'] = main_df['Credit_History'].fillna(main_df['Credit_History'].mean())
```

```python
main_df['LoanAmount'].median()
```

```python
main_df['LoanAmount'].mean()
```

```python
# Martial status 
main_df['Gender'] = main_df['Gender'].fillna(main_df['Gender'].mode()[0])
main_df['Married'] = main_df['Married'].fillna(main_df['Married'].mode()[0])
main_df['Dependents'] = main_df['Dependents'].fillna(main_df['Dependents'].mode()[0])
 
# self-employ jobs
main_df['Self_Employed'] = main_df['Self_Employed'].fillna(main_df['Self_Employed'].mode()[0])
```

```python
print("number of peoples with loan approval: ", main_df['Loan_Status'].value_counts()[0])
print("number of peoples with loan rejection: ", main_df['Loan_Status'].value_counts()[1])
plt.figure(figsize=(6, 4))
sns.countplot(x='Loan_Status', data=main_df)
```
